```python
# All imports
from sklearn.datasets import fetch_olivetti_faces
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPClassifier
from sklearn.preprocessing import StandardScaler
from sklearn import metrics
import seaborn as sns
import matplotlib.pyplot as plt
from datetime import datetime
```

```python
# Importing california housing data set

housingData = fetch_olivetti_faces()
X = housingData.data
Y = housingData.target
```

```python
# Splitting dataset into 70% training and 30% testing

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, train_size=0.7, random_state=1)
```

```python
#Preporcessing the data

stdsc = StandardScaler()
stdsc.fit(X_train)
X_train_std = stdsc.transform(X_train)
X_test_std = stdsc.transform(X_test)
```

```python
NO_OF_ITERATIONS = 100
```

```python
Y_train=Y_train.astype('int')
Y_test=Y_test.astype('int')
```

```python
# NN 1 - No of hidden layers = 2; No of neurons per hidden layer = 10; Activation function = relu

startTime = datetime.now()
HIDDEN_LAYER_PARAMETERS = [2, 10]
mlp = MLPClassifier(max_iter=NO_OF_ITERATIONS, random_state=1, activation='relu', hidden_layer_sizes=HIDDEN_LAYER_PARAMETERS)
mlp.fit(X_train_std, Y_train)

print("Total time taken to train: {}".format(datetime.now() - startTime))
print("Activation function: {}".format(mlp.activation))
print("List of predicted Classes: {}".format(mlp.classes_))
print("Training set loss: {}".format(mlp.loss_))
print("Accuracy score: {}".format(mlp.score(X_test_std, Y_test)))
print("Output activation function: {}".format(mlp.out_activation_))
# evaluation matrix
Y_predicted = mlp.predict(X_test_std)
print(metrics.classification_report(Y_test, Y_predicted))
# Confusion Matrix
mat = metrics.confusion_matrix(Y_test, Y_predicted)
sns.heatmap(mat.T, square=True, annot=True, fmt='d',cbar=False)
plt.xlabel("True label")
plt.ylabel("Predicted label")
```
